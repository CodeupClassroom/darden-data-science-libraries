# Matplotlib

```python
import matplotlib.pyplot as plt
```

## Basic Plots

- `plt.plot`
    - with a single arg
    - with multiple series
- `plt.scatter`
- lines + points together

```python
y = [2, 1, 3]

plt.plot(y)
```

```python
x = [3, 2, 1]

plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.scatter(x, y)
```

```python
x = [1, 2, 3]
y = [4, 5, 6, 7]

plt.scatter(x, y)
```

## Customizing Visuals

kwargs for appearance customization: `c`, `s`, `alpha`, `ls`, `marker`

- `c` or `color`
- `s` for size of scatter plot points
- `alpha` to adjust transparency
- `ls` to adjust the linestyle -- `-` for dashed, `:` for dotted

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, c='red', ls=':')
plt.scatter(x, y, s=500, alpha=.5)
```

## labels + titles

tangent: LaTeX

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, c='red', ls=':')
plt.scatter(x, y, s=500, alpha=.5)
plt.title('My Awesome Viz: $y = x^2$')
plt.xlabel('The x axis')
plt.ylabel('The y axis')
```

## axis ticks: custom ticks, rotation

```python
y = [2, 3, 1, 2, 5]

plt.plot(y)
plt.xticks([0, 10], rotation=180)
```

## `.text` + `.annotate`

```python
x = ['Ada', 'Bayes', 'Curie', 'Darden']
y = [17, 20, 16, 21]

plt.bar(x, y, color='hotpink')
plt.yticks(range(0, 22, 2))
plt.title('# Of Students by Data Science Cohort')

# (x, y, text)
# ha == horizontal align
plt.text(0, 19, "First Cohort", ha='center')
plt.annotate('Second Cohort', xy=(1, 20), xytext=(2, 18), arrowprops={'arrowstyle': '->'})
```

Legends + Figure Size

```python
plt.figure(figsize=(16, 9))

x = [1, 2, 3, 4, 5]
y1 = [1, 1, 1, 3, 4]
y2 = [0, 2, 3, 3, 3]

plt.plot(x, y1, label='Python Knowledge')
plt.plot(x, y2, label='SQL Knowledge')
plt.legend()
```

## histograms

```python
import random

# 50 random numbers between 0 and 1
x = [random.random() for _ in range(500)]

plt.hist(x, bins=25)
plt.title('Histogram of 500 random numbers')
```

## subplots

twiny

```python
# 50 random numbers betwee 0 and 100
x = [random.random() * 100 for _ in range(50)]

# y1 is a function of x + some random noise
y1 = [n + random.random() * 10 for n in x]
y2 = [100 - n + random.random() * 10 for n in x]
```

```python
plt.scatter(x, y1)
plt.scatter(x, y2)
```

```python
# plt.subplot(nrows, ncols, current axis)
# the number of rows and columns should be the same for each call to subplot within a cell

plt.subplot(2, 1, 1)
plt.scatter(x, y1)

plt.subplot(2, 1, 2)
plt.scatter(x, y2)
```

```python
plt.subplot(1, 2, 1)
plt.hist(y1)
plt.title('histogram of y1')

plt.subplot(1, 2, 2)
plt.scatter(x, y1)
plt.title('x vs y1')
```

```python
plt.figure(figsize=(16, 9))

plt.subplot(2, 3, 1)
plt.hist(x, color='limegreen')
plt.title('histogram of x')

plt.subplot(2, 3, 2)
plt.hist(y1, color='firebrick')
plt.title('histogram of y1')

plt.subplot(2, 3, 3)
plt.hist(y2, color='rebeccapurple')
plt.title('histogram of y2')

plt.subplot(2, 3, 4)
plt.plot(x, color='burlywood')
plt.title('Plot of x')

plt.subplot(2, 3, 5)
plt.scatter(x, y1, color='khaki')
plt.title('x vs y1')

plt.subplot(2, 3, 6)
plt.scatter(x, y2, color='royalblue')
plt.title('x vs y2')
```

Show off a fun visualization with funky fonts

```python
time = [1, 2, 3, 4, 5, 6, 7, 8, 9]
python_knowledge = [1, 1, 1, 1, 1, 3, 4, 5, 6]
sql_knowledge = [0, 2, 3, 4, 5, 5, 5, 5, 5]
confidence = [2, 3, -1, -4, 10, 1, -10, 6, 4]

plt.figure(figsize=(16, 9))

plt.plot(time, python_knowledge, label='python')
plt.plot(time, sql_knowledge, label='SQL')
plt.plot(time, confidence, label='Confidence')
plt.legend()

plt.title('My Experience at Codeup')
plt.xlabel('Time')
plt.annotate('Started Python', xy=(5, 0), xytext=(2, -4), arrowprops={'arrowstyle': '->'})
```
